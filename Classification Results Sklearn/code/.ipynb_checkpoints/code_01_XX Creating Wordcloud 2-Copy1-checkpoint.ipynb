## 01_02 Preparing Data for Wordcloud

```python
import pandas as pd
data = pd.read_csv("questions.csv")
data.head(50)
```

```python
data.columns = ["words"]
```

```python
words = data["words"].str.cat(sep=".")
```

```python
import os
import codecs
from wordcloud import WordCloud
import arabic_reshaper
from bidi.algorithm import get_display

# Make text readable for a non-Arabic library like wordcloud
text = arabic_reshaper.reshape(words)
text = get_display(text)

# Generate a word cloud image
wordcloud = WordCloud(font_path='NotoSansArabicUI-Regular.ttf',background_color="white",width=1920,height=1080, max_words=50).generate(text)

# Export to an image
wordcloud.to_file("arabic_example.png")
```

## 01_03 Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
