## 01_02 Preparing Data for Wordcloud

```python
import pandas as pd
data = pd.read_csv("appen_translated - hotels.csv")
data.head(50)
```

```python
words = data["review_en"].str.cat(sep=".")
```

```python
words[:10000]
```

```python
import os
import codecs
from wordcloud import WordCloud

# Generate a word cloud image
wordcloud = WordCloud(background_color="white",width=1080,height=720, max_words=50).generate(words)

# Export to an image
#wordcloud.to_file("arabic_example.png")
```

```python
words_arr = wordcloud.words_
```

```python
words_arr.keys()
```

## 01_03 Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud)
mpLib.axis("off")
mpLib.show()
```
