## 01_02 Preparing Data for Wordcloud

```python
import pandas as pd
data = pd.read_csv("appen_translated - books.csv")
data.head(50)
```

```python
reveiws_help = data['review_en'][data["helpfulness"] == "helpful"].str.cat(sep=".")
reveiws_non = data['review_en'][data["helpfulness"] != "helpful"].str.cat(sep=".")
```

```python
len(reveiws_help)
```

```python
len(reveiws_non)
```

```python
#to install wordcloud - execute "pip install wordcloud" from Anaconda prompt

#read the course descriptions
#with open("Course-Descriptions.txt", 'r') as fh:  
#    filedata = fh.read()

#check contents
#print("File data sample : ", filedata[:200])

# Create stopword list:
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)

#read text


#Generate wordcloud data
wordcloud_h = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(reveiws_help)

wordcloud_n = WordCloud(stopwords=stopwords, max_words=25, \
                      background_color="white").generate(reveiws_non)
```

```python
print(wordcloud_h.words_.keys())
```

```python
print(wordcloud_n.words_.keys())
```

## 01_03 Displaying the Wordcloud

```python
import matplotlib.pyplot as mpLib
mpLib.imshow(wordcloud_h)
mpLib.axis("off")
mpLib.show()
```

```python
mpLib.imshow(wordcloud_n)
mpLib.axis("off")
mpLib.show()
```
