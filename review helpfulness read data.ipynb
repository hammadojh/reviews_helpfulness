```python
from os import listdir
from os.path import isfile, join
onlyfiles = [f for f in listdir('acl_data/reviews/book') if isfile(join('acl_data/reviews/book', f))]

# init dict
books_dict = dict()
for f in onlyfiles:
    books_dict[f[:f.index('.')]] = {'product':'','review':''}
    
# add files content to dict
for f in onlyfiles:

    file = open("acl_data/reviews/book/{}".format(f),"r")
    text = file.read()
    file.close()
    
    if ".product" in f:
        books_dict[f[:f.index('.')]]['product'] = text
    else:
        books_dict[f[:f.index('.')]]['review'] = text
    
```

```python
scores = [[key,value['review'][value['review'].index(" ")+1:value['review'].index("/")-2]] for key,value in books_dict.items()]
scores = pd.DataFrame(scores)
scores.columns = ['name','rating']
scores
```

```python
merged = merged.merge(scores, left_on='name', right_on='name')
merged
```

```python
reviews = [[key,value['review'][value['review'].index("Summary")+9:]] for key,value in books_dict.items()]
reviews = pd.DataFrame(reviews)
reviews.columns = ['name','review']
reviews
```

```python
merged = merged.merge(reviews, left_on='name', right_on='name')
merged
```
